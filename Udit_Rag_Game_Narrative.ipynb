```python
def get_ai_tags(text_to_analyze, source_type):
    """Sends text to the Gemini API to get interpretive tags."""
    meta_prompt = f"""
    You are an expert librarian and data analyst specializing in game narrative design. I am providing you with text from a document of type: **{source_type}**.
    Based on the content of this **{source_type}**, analyze it and return a single, minified JSON object with your suggested tags for the following fields: `primary_genre`, `key_topics` (list of 3-5), and `narrative_blueprint`. Adhere strictly to the provided tag options.

    Pre-defined Tag Options:
    * `primary_genre`: Choose one from `['RPG', 'Horror', 'Adventure', 'Strategy', 'Simulation', 'General']`.
    * `key_topics`: Choose up to five from `['World-Building', 'Quest Design', 'Character Arc', 'Dialogue', 'Game Mechanics', 'Emotional Impact', 'Narrative Theory', 'Postmortem']`.
    * `narrative_blueprint`: Choose one from `['Hero\\'s Journey', 'Open World', 'Linear Narrative', 'Three-Act Structure', 'No Blueprint']`.

    Text from {source_type} to Analyze:
    \"\"\"{text_to_analyze}\"\"\"

    JSON Output:
    """
    try:
        response = model.generate_content(meta_prompt)
        # Clean the response to ensure it's valid JSON
        cleaned_response = response.text.strip().replace('`', '')
        if cleaned_response.startswith('json'):
            cleaned_response = cleaned_response[4:]
        return json.loads(cleaned_response)
    except Exception as e:
        print(f"--> ERROR calling Gemini API or parsing JSON: {e}")
        return None
```

```python
def normalize_whitespace(text):
    """Collapses multiple whitespace characters into a single space."""
    text = re.sub(r'\s+', ' ', text)
    text = text.strip()
    return text

def remove_urls(text):
    """Removes URLs from the text."""
    # A simple regex for URLs
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    return url_pattern.sub(r'', text)

def remove_screenplay_artifacts(text):
    """Removes artifacts common to screenplays and scripts."""
    # Remove scene headings (e.g., INT. ROOM - DAY)
    text = re.sub(r'^(INT\.|EXT\.)[^\n]+\n', '', text, flags=re.MULTILINE)
    # Remove character cues in all caps
    text = re.sub(r'^[A-Z\s]+\n(?=[A-Z])', '', text, flags=re.MULTILINE)
    # Remove parenthetical directions (e.g., (V.O.), (beat))
    text = re.sub(r'\([^\)]+\)', '', text)
    # Remove page numbers
    text = re.sub(r'^\d+\.?\s*$', '', text, flags=re.MULTILINE)
    return text

def remove_transcript_artifacts(text):
    """Removes artifacts common to video transcripts."""
    # Remove timestamps (e.g., [00:25:10])
    text = re.sub(r'\[\d{2}:\d{2}:\d{2}\]', '', text)
    # Remove speaker labels (e.g., Speaker 1:)
    text = re.sub(r'^[A-Za-z0-9\s]+:\s', '', text, flags=re.MULTILINE)
    # Remove sounds in brackets (e.g., [Applause])
    text = re.sub(r'\[[^\]]+\]', '', text)
    return text
```

```python
def clean_text(text, source_type):
    """
    Main controller function to clean text based on its source type.
    """
    # Stage 1: Apply universal normalization to all text
    text = ftfy.fix_text(text) # Fixes unicode and encoding errors
    text = remove_urls(text)

    # Stage 2: Apply format-specific cleaning
    if source_type in ['Screenplay', 'Script']:
        text = remove_screenplay_artifacts(text)
    elif source_type == 'Video Essay Transcript':
        text = remove_transcript_artifacts(text)

    # Stage 3: Final whitespace normalization
    text = normalize_whitespace(text)

    return text
```

```python

print("--- Starting Batch Processing ---")
for file_info in file_list:
    print(f"\nProcessing file: {file_info['source_title']}...")

    # STAGE 1: Text Extraction
    full_text = extract_text_from_pdf(file_info['pdf_path'])
    if not full_text:
        continue # Skip to next file if extraction failed

    # STAGE 2: Content Sampling (if needed)
    text_for_analysis = ""
    ## print length of the text
    print("Text length:  ", len(full_text))
    # Approx. 700k token limit (1 token ~= 4 chars)
    if len(full_text) > 2800000:
        start = full_text[:700000]
        middle = full_text[len(full_text)//2 : len(full_text)//2 + 700000]
        end = full_text[-700000:]
        text_for_analysis = f"{start}\n...[CONTENT SNIPPED]...\n{middle}\n...[CONTENT SNIPPED]...\n{end}"
    else:
        text_for_analysis = full_text

    # STAGE 3: AI-Assisted Tagging
    print("text length Post Approx: ", len(text_for_analysis))
    clean_text_for_analysis = clean_text(text_for_analysis, file_info['source_type'])
    print("Clean text length: ", len(clean_text_for_analysis))
    # print(text_for_analysis)
    ai_suggested_tags = get_ai_tags(clean_text_for_analysis, file_info['source_type'])
    if not ai_suggested_tags:
        print(f"--> Skipping file due to tagging error.")
        continue
    print(f"--> AI Suggested Tags: {ai_suggested_tags}")

    # STAGE 4: Curation & Final Metadata Assembly
    # For this script, we auto-approve the AI tags. In a real workflow, you might add a manual review step here.
    final_metadata = {
        "source_title": file_info['source_title'],
        "source_type": file_info['source_type'],
        **ai_suggested_tags # Neatly combines the two dictionaries
    }
    final_metadata_json = json.dumps(final_metadata)
    print(f"--> Final Metadata: {final_metadata_json}")
    # STAGE 5: Chunking & Saving
    chunks = text_splitter.split_text(full_text)
    with open(output_filepath, 'a', encoding='utf-8') as f:
        for i, chunk in enumerate(chunks):
            record = {
                "content_id": f"{file_info['source_title'].replace(' ', '-').lower()[:20]}_{i}",
                "extraction_date_utc": datetime.utcnow().isoformat(),
                "raw_text": chunk,
                **final_metadata # Add all metadata to each chunk
            }
            f.write(json.dumps(record) + '\n')
    time.sleep(2) # Be nice to the API
    print(f"--> ✅ Successfully processed and saved {len(chunks)} chunks.")
```

```python
## load the dataset and intialise it in the vector db

from langchain_chroma import Chroma
from langchain.schema import Document

# Define the path for the persistent database
persist_directory = "/content/AI_Mentor_Project/db/"

# --- Prepare documents for ChromaDB ---
# ChromaDB works best with LangChain's Document objects.
# We will convert our list of dictionaries into a list of Document objects.
# Each Document needs a 'page_content' and a 'metadata' dictionary.

langchain_documents = []
for doc_dict in documents:
    page_content = doc_dict.get("raw_text", "")

    # Build metadata by copying and excluding 'raw_text'
    metadata = {k: v for k, v in doc_dict.items() if k != "raw_text"}

    if isinstance(metadata.get("key_topics"), list):
        metadata["key_topics"] = ", ".join(metadata["key_topics"])

    langchain_documents.append(Document(page_content=page_content, metadata=metadata))

print(langchain_documents[0])


```

```python
# --- Create and Populate the ChromaDB Database ---
try:
    print("🚀 Starting to create the vector database... This will take several minutes.")

    # This single command creates the embeddings and stores them in the database.
    # It will use the 'embeddings' function we initialized in the previous step.
    vectordb = Chroma.from_documents(
        documents=langchain_documents,
        embedding=embeddings,
        persist_directory=persist_directory
    )

    # Optional: You can confirm the number of items in the database
    print(f"\n✅ Vector database created successfully with {vectordb._collection.count()} documents.")
    print(f"Database is persisted at: {persist_directory}")

except Exception as e:
    print(f"An error occurred during database creation: {e}")
```

```python
# --- 1. Define a Test User Query ---
# This simulates the inputs we would get from our final application's UI.
test_idea = "A lone astronaut on a derelict spaceship who discovers the crew vanished due to a psychological entity."
test_genre = "Adventure"
test_blueprint = "Open World"
test_topic = "World-Building"

print(f"--- Test Query ---")
print(f"Idea: '{test_idea}'")
print(f"Genre: '{test_genre}' | Blueprint: '{test_blueprint}' | Topic: '{test_topic}' \n")


```

```python
# --- 2. Implement the Two-Stage Retrieval Strategy ---

# Stage 1: Create the Metadata Pre-Filter
# This 'where' clause tells ChromaDB to only search documents that match BOTH conditions.
metadata_filter = {
    "$and": [
        {"primary_genre": {"$eq": test_genre}},
        {"narrative_blueprint": {"$eq": test_blueprint}},
    ]
}

# Add conditional logic based on topic
if test_topic == "World-Building":
    metadata_filter["$and"].append(
        {
            "$or": [
                {"source_type": {"$eq": "blog"}},
                {"source_type": {"$eq": "Video Essay Transcript"}}
            ]
        }
        )

metadata_filter

```

```python
# Stage 2: Configure the Retriever
# We create a retriever from our loaded vector database 'vectordb'.
retriever = vectordb.as_retriever(
    search_type="mmr", # Use Maximal Marginal Relevance to get diverse results
    search_kwargs={
        'k': 4,                  # Return the top 4 most relevant documents
        'fetch_k': 10,           # Initially fetch 20 documents to run MMR on
        'filter': metadata_filter  # Apply our metadata pre-filter
    }
)
retriever


# # --- 3. Execute the Query ---
# # The retriever now performs the full two-stage search.
retrieved_docs = retriever.invoke(test_idea)

# After getting results
filtered_results = [
    doc for doc in retrieved_docs
    if test_topic in doc.metadata.get("key_topics", "")
]



# retrieved_docs

# --- 4. Display the Results ---
print(f"--- Retrieved Documents ({len(retrieved_docs)} found) ---")
print(f"--- Filtered Documents ({len(filtered_results)} found) ---")
for i, doc in enumerate(filtered_results):
    print(f"\n--- Document {i+1} ---\n")
    # Print the metadata associated with that chunk
    print("\nMETADATA:")
    print(doc.metadata)
    # Print the text content of the chunk
    print("CHUNK TEXT:")
    print(doc.page_content)
```
